```python
data = pd.read_csv('train.csv')
print(data.shape)
data.head()
```

```python
data_clean = data.dropna()
```

```python
m = 10
c = 10

#h = mx + c
totalError = 0
for index, row in data_clean.iterrows():
    x, y = row.x, row.y
    error = ((m*x+c)-y)**2
    totalError += error
    
print(len(data_clean))
avgError = totalError / len(data_clean)
avgError
```

```python
startAvgError = prevAvgError = avgError

N = 0
while prevAvgError >= avgError:
    mDerivative = 0
    cDerivative = 0
    for index, row in data_clean.iterrows():
        x, y = row.x, row.y
        mDerivative += 0.0002*(2*(m*x+c-y)*x)/len(data_clean)
        cDerivative += 0.0002*(2*(m*x+c-y))/len(data_clean)

    prevM, prevC = m, c
    m -= mDerivative
    c -= cDerivative

    totalError = 0
    for index, row in data_clean.iterrows():
        x, y = row.x, row.y
        error = ((m*x+c)-y)**2
        totalError += error

    prevAvgError = avgError
    avgError = totalError / len(data_clean)
    N += 1
    print(N, avgError, prevAvgError-avgError, startAvgError-avgError)
    
N, avgError, prevAvgError
```
