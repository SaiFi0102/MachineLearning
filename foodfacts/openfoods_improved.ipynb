**Function from https://stackoverflow.com/questions/12680754/split-explode-pandas-dataframe-string-entry-to-separate-rows**

```python
def tidy_split(df, column, sep='|', keep=False):
    """
    Split the values of a column and expand so the new DataFrame has one split
    value per row. Filters rows where the column is missing.

    Params
    ------
    df : pandas.DataFrame
        dataframe with the column to split and expand
    column : str
        the column to split and expand
    sep : str
        the string used to split the column's values
    keep : bool
        whether to retain the presplit value as it's own row

    Returns
    -------
    pandas.DataFrame
        Returns a dataframe with the same columns as `df`.
    """
    indexes = list()
    new_values = list()
    df = df.dropna(subset=[column])
    for i, presplit in enumerate(df[column].astype(str)):
        values = presplit.split(sep)
        if keep and len(values) > 1:
            indexes.append(i)
            new_values.append(presplit)
        for value in values:
            indexes.append(i)
            new_values.append(value)
    new_df = df.iloc[indexes, :].copy()
    new_df[column] = new_values
    return new_df
```

### Get unique countries and entries according to countries

```python
countries = data_validcols[['countries_en']].dropna()
countries = tidy_split(countries, column='countries_en', sep=',')
countries = countries.groupby(['countries_en']).size()
countries = pd.DataFrame(countries.sort_values(ascending=False))
countries.columns = ["Entries"]
countries
```

```python
countries_bar = countries[0:20]
countries_other = countries[20:].agg(['sum'])
countries_bar = countries_bar.append(countries_other)
countries_bar = countries_bar.rename({'sum':'Other'})
countries_bar.plot(kind="bar", figsize=(15, 8))
```

### Palm oil examination

```python
palmoilcols = ['ingredients_from_palm_oil_n', 'ingredients_that_may_be_from_palm_oil_n']
countrycols = ['countries_en']
bothcols = palmoilcols + countrycols
data_palmoil = data_validcols[bothcols]
data_palmoil = data_palmoil.dropna(subset=palmoilcols)
data_palmoil['palm_c'] = data_palmoil[palmoilcols].max(axis=1)

other_thres = 15
data_palmoil_t = data_palmoil[data_palmoil['palm_c'] != 0]
data_palmoil_t = tidy_split(data_palmoil_t, column='countries_en', sep=',')
data_palmoil_tcc = pd.DataFrame(data_palmoil_t.groupby(['countries_en']).size().sort_values(ascending=False))
tcc_other = data_palmoil_tcc[other_thres:].agg(['sum'])
tcc_other = tcc_other.rename({"sum":"Other"})
data_palmoil_tcc = data_palmoil_tcc.drop(data_palmoil_tcc.index[other_thres:])
data_palmoil_tcc = data_palmoil_tcc.append(tcc_other)

data_palmoil_f = data_palmoil[data_palmoil['palm_c'] == 0]
data_palmoil_f = tidy_split(data_palmoil_f, column='countries_en', sep=',')
data_palmoil_fcc = pd.DataFrame(data_palmoil_f.groupby(['countries_en']).size().sort_values(ascending=False))
fcc_other = data_palmoil_fcc[other_thres:].agg(['sum'])
fcc_other = fcc_other.rename({"sum":"Other"})
data_palmoil_fcc = data_palmoil_fcc.drop(data_palmoil_fcc.index[other_thres:])
data_palmoil_fcc = data_palmoil_fcc.append(fcc_other)

data_palmoil_bar = pd.concat([data_palmoil_tcc, data_palmoil_fcc], axis=1)
data_palmoil_bar.columns = ["Items with palmoil", "Items without palmoil"]
data_palmoil_bar.plot(kind='bar', figsize=(15,5))
```

### Food Categories

```python
categories = data_validcols[['categories','categories_tags','categories_en']].dropna()
categories = tidy_split(categories, column='categories_tags', sep=',')
categories = pd.DataFrame(categories.groupby(['categories_tags']).size().sort_values(ascending=False))
categories = categories
categories.columns = ['Count']
categories
```

```python
data_cats = data_validcols.dropna(subset=['categories_tags'])

cols_base = ['categories_tags']
cols_good = [
'energy_100g',
'fat_100g',
'saturated-fat_100g',
'carbohydrates_100g',
'sugars_100g',
'fiber_100g',
'proteins_100g',
'salt_100g',
'sodium_100g'
]
cols_bad = [
'vitamin-a_100g',
'vitamin-c_100g',
'calcium_100g',
'iron_100g',
'nutrition-score-fr_100g',
'nutrition-score-uk_100g'
]

data_nutr = []
data_vege = []
data_nonv = []
for col in cols_good + cols_bad:
    data_nutr.append(data_cats[cols_base + [col]].dropna())
    data_vege.append(data_nutr[-1][data_nutr[-1].categories_tags.str.contains("vegan|vegetarian")])
    data_nonv.append(data_nutr[-1][~data_nutr[-1].categories_tags.str.contains("vegan|vegetarian")])
    if(data_vege[-1].shape[0] < 10):
        data_nutr.pop()
        data_vege.pop()
        data_nonv.pop()
    print(col + ": (" + str(data_vege[-1].shape[0]) + ", " + str(data_nonv[-1].shape[0]) + ")")
```

```python
nutr_vege = []
nutr_nonv = []
for df in data_vege: nutr_vege.append(df.mean())
for df in data_nonv: nutr_nonv.append(df.mean())
```

```python
nutrs = []
energy = []
for i in range(len(nutr_vege)):
    nutrs.append(pd.concat([nutr_vege[i], nutr_nonv[i]], axis=1))
    nutrs[-1].columns = ["Vegan", "Non vegan"]

    energy.append(nutrs[-1].loc['energy_100g'])
    nutrs[-1] = nutrs[-1].drop(['energy_100g'])
```

```python
ax = nutrs[0].plot(kind='bar', figsize=(10,5))
for i in range(1, len(nutrs)):
    nutrs[i].plot(kind='bar', ax=ax)
```
