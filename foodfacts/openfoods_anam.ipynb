**Function from https://stackoverflow.com/questions/12680754/split-explode-pandas-dataframe-string-entry-to-separate-rows**

```python
def tidy_split(df, column, sep='|', keep=False):
    """
    Split the values of a column and expand so the new DataFrame has one split
    value per row. Filters rows where the column is missing.

    Params
    ------
    df : pandas.DataFrame
        dataframe with the column to split and expand
    column : str
        the column to split and expand
    sep : str
        the string used to split the column's values
    keep : bool
        whether to retain the presplit value as it's own row

    Returns
    -------
    pandas.DataFrame
        Returns a dataframe with the same columns as `df`.
    """
    indexes = list()
    new_values = list()
    df = df.dropna(subset=[column])
    for i, presplit in enumerate(df[column].astype(str)):
        values = presplit.split(sep)
        if keep and len(values) > 1:
            indexes.append(i)
            new_values.append(presplit)
        for value in values:
            indexes.append(i)
            new_values.append(value)
    new_df = df.iloc[indexes, :].copy()
    new_df[column] = new_values
    return new_df
```

### Get unique countries

```python
countries = data_validcols[['countries','countries_tags','countries_en']]
countries = tidy_split(countries, column='countries_en', sep=',')

countries_unique = countries.groupby(['countries_en']).size()
countries_unique = pd.DataFrame(countries_unique.sort_values(ascending=False))
countries_unique
```

### Palm oil examination

```python
palmoilcols = ['ingredients_from_palm_oil_n', 'ingredients_that_may_be_from_palm_oil_n']
countrycols = ['countries_en']
bothcols = palmoilcols + countrycols
data_palmoil = data_validcols[bothcols]
data_palmoil = data_palmoil.dropna(subset=palmoilcols)
data_palmoil['palm_c'] = data_palmoil[palmoilcols].max(axis=1)

data_palmoil_t = data_palmoil[data_palmoil['palm_c'] == 0]
data_palmoil_t = tidy_split(data_palmoil_t, column='countries_en', sep=',')
data_palmoil_tcc = pd.DataFrame(data_palmoil_t.groupby(['countries_en']).size().sort_values(ascending=False))

data_palmoil_f = data_palmoil[data_palmoil['palm_c'] != 0]
data_palmoil_f = tidy_split(data_palmoil_f, column='countries_en', sep=',')
data_palmoil_fcc = pd.DataFrame(data_palmoil_f.groupby(['countries_en']).size().sort_values(ascending=False))

data_palmoil_tcc.shape, data_palmoil_fcc.shape
```
